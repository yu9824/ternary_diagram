```python
from ternary_diagram import TernaryDiagram
import pandas as pd
```

```python
df = pd.read_csv('example_contour.csv')
```

Compounds corresponding to vertices

```python
materials = df.columns[0:3]
```

```python
td = TernaryDiagram(materials=materials)  # 
td.contour(df[materials], z = df['z'], minimum = 0, maximum = 1, bar_label = 'example_contour', norm = True)
```

```python
td.fig.savefig('example_contour.png', dpi = 300)
```
